```python
#1 dataset

import os
import io
import pandas as pd

df = pd.read_csv("DATA.csv")

x = df["x"]
y = df["y"]
print(x,y)
```

```python
#2 parameter initialization

theta = [1,2]
print(theta[0],theta[1])
```

```python
count = 0
for i in range(1000): 
    #3 hypothesis function
    print("iteration number:", count)
    count = count + 1 
    
    m = len(y)
    h = []

    for i in range(m):
        h.append(theta[0] + theta[1]*x[i])

    #4 cost function

    error = 0
    for i in range(m):
        error = error + (h[i] - y[i])**2

    J = (1/(2*m))*error
    print("cost is :", J)


    #5 gradient descent
    sum0 = 0
    sum1 = 0

    for i in range(m):
        sum0 = sum0 + (h[i] - y[i])
        sum1 = sum1 + (h[i] - y[i])*x[i]

    alpha = 0.001
    theta[0] = theta[0] - (alpha/m)*sum0
    theta[1] = theta[1] - (alpha/m)*sum1

    print("theta[0] is :", theta[0], "theta[1] is :", theta[1])
```
