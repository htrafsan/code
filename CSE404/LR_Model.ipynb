```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array([2,4,6,7]).reshape(-1,1)
y = np.array([3,5,8,10])

model = LinearRegression().fit(x,y)

print('intercept/theta0:', model.intercept_)
print('slope/theta1:', model.coef_)

y_predic = model.intercept_ + model.coef_ * x

import matplotlib.pyplot as plt
plt.plot(x,y, marker="*")
plt.plot(x,y_predic, marker="s")
```
